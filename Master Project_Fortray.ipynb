```python
invalid_time = df["Tap_Out_Time"] < df["Tap_In_Time"]

df[invalid_time]
```

```python
quarantine_df = df[invalid_time]
clean_df = df[~invalid_time]
```

```python
df["Tap_In_Time"] = pd.to_datetime(df["Tap_In_Time"])
df["Tap_Out_Time"] = pd.to_datetime(df["Tap_Out_Time"])
```

```python
df["Journey_Duration"] = (
    df["Tap_Out_Time"] - df["Tap_In_Time"]
).dt.total_seconds() / 60
```

```python
impossible_duration = df["Journey_Duration"] < 1
```

```python
quarantine_df = df[impossible_duration].copy()
clean_df = df[~impossible_duration].copy()
```

```python
quarantine_df.to_csv("tfl_quarantined_data.csv", index=False)
```

```python
clean_df["Total_Journey_Duration"] = (
    clean_df["Tap_Out_Time"] - clean_df["Tap_In_Time"]
).dt.total_seconds() / 60
```

```python
clean_df["Cost_Per_Minute"] = (
    clean_df["Journey_Cost"] / clean_df["Total_Journey_Duration"]
)
```

```python
clean_df["Hour"] = clean_df["Tap_In_Time"].dt.hour
```

```python
def get_time_of_day(hour):
    if 7 <= hour < 10:
        return "Peak AM"
    elif 17 <= hour < 20:
        return "Peak PM"
    else:
        return "Off Peak"
```

```python
clean_df["Time_of_Day"] = clean_df["Hour"].apply(get_time_of_day)
```

```python
def congestion_engine(row):
    if (
        row["Time_of_Day"] == "Peak AM"
        and row["Delay_Minutes"] > 15
        and row["Total_Journey_Duration"] > 60
    ):
        return "Severe Friction"
    return "Smooth"

clean_df["Congestion_Tier"] = clean_df.apply(congestion_engine, axis=1)
```

```python
clean_df.to_csv("tfl_commuter_trends.csv", index=False)
```

```python
import sqlite3

# create database file
conn = sqlite3.connect("tfl_data.db")

# save cleaned dataset into SQL table
clean_df.to_sql("journeys", conn, if_exists="replace", index=False)

conn.close()
```
